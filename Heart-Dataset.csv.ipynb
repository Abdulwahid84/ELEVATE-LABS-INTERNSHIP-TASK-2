```python
def remove_outliers(df1, categorical_cols):
    
    df_cleaned = df1.copy()

    for i in categorical_cols:
        
        Q1 = df_cleaned[i].quantile(0.25)
        Q3 = df_cleaned[i].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        
        df_cleaned = df_cleaned[(df_cleaned[i] >= lower_bound) & (df_cleaned[i] <= upper_bound)]
    
    return df_cleaned

numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

df_no_outliers = remove_outliers(df1, categorical_cols)


print(f"Original DataFrame shape: {df1.shape}")
print(f"Cleaned DataFrame shape: {df_no_outliers.shape}")
```
